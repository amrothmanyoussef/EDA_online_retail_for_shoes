## Task2: Data Cleaning and Preprocessing

```python
#Remove duplicate rows if exists
df = df.drop_duplicates()
df
```

Number of rows is decreased from 541909 to 536641 meaning that there were 5268 duplicate rows removed

```python
#Show number of missing elements in each column
df.isna().sum()
```

```python
#fill missing data with pd.NA to be able to treat with our data and convert data types
df = df.fillna(pd.NA)
```

```python
CustomerID_NA_index=df[df.CustomerID.isna()].index
```

'CustomerID' column data type is float while it is Categorical, so it must be converted to string

```python
#convert 'CustomerID' column to string
df['CustomerID'] = df['CustomerID'].astype(str)
df['CustomerID']
```

```python
df.isna().sum()
```

```python
pd.DataFrame(df,
            index=CustomerID_NA_index)
```

```python
df['CustomerID'] = np.where(df['CustomerID']=='<NA>',
                            pd.NA,
                            df['CustomerID'])
```

```python
df.isna().sum()
```

```python
#remove extra zero and decimal point '.0' 
df['CustomerID'] = df['CustomerID'].str.strip("0") #remove zero
df['CustomerID']
```

```python
#remove extra zero and decimal point '.0' 
df['CustomerID'] = df['CustomerID'].str.strip(".") #remove decimal point
df['CustomerID']
```

```python
df.isna().sum()
```

```python
#Convert 'InvoiceDate' to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.dtypes
```

```python
#Check Country names
df['Country'].unique()
```

Replace 'Unspecified' countries with pd.NA to visualize correctly

```python
df['Country'] = np.where(df['Country'] == 'Unspecified',
                        pd.NA,
                        df['Country'])
df['Country'].unique()
```

```python
df.isna().sum()
```

```python
#Remove Description column since no need for it in statistics or visualization
df = df.drop("Description",axis=1)
df
```

Detect Outlier with box plot for numerical data columns

```python
sns.boxplot(x=df['Quantity'])
```

```python
sns.boxplot(x=df['UnitPrice'])
```

There are outliers in both columns.

lets see how they affect statistics:

```python
df.describe()
```

```python
#Seperate numerical data in another DAtaFrame to remove outliers easily
numeric=['Quantity','UnitPrice']
df2=df[numeric]
df2
```

```python
Q1 = df2.quantile(0.25)
Q3 = df2.quantile(0.75)
IQR = Q3 - Q1 # Interquartile Range
print(IQR)
```

```python
#Remove Outliers
df2 = df2[~((df2 < (Q1 - 1.5 * IQR)) |(df2 > (Q3 + 1.5 * IQR))).any(axis=1)]
df2.shape
```

Number of rows is decreased from 536641 to 439135 meaning that there are 97506 outlier!

Lets compare between statistic before removing outliers and after:

```python
#Before removing outliers
df.describe()
```

```python
#After removing outliers
df2.describe()
```

```python
sns.boxplot(x=df2['UnitPrice'])
```

```python
df2.head()
```

```python
#Group df2 by 'CustomerID' to Extract important insights
column_of_interest=['CustomerID','UnitPrice','Quantity']
by_CustomerID = df2[column_of_interest].groupby('CustomerID').sum()
by_CustomerID
```

```python
#Change column 'UnitPrice' to 'Revenue$'
by_CustomerID = by_CustomerID.rename(columns={'UnitPrice':'Revenue$'})
by_CustomerID
```

```python
#sort DataFrame by Revenue to get data to get the data of customer who pays most
by_CustomerID = by_CustomerID.sort_values('Revenue$',ascending=False)
by_CustomerID
```

```python
print("highest Payment for a customer is",by_CustomerID.iloc[0].loc['Revenue$'],'$')
by_CustomerID.iloc[0]
```

```python
#sort DataFrame by Quantity to get data to get the data of customer who buy highest number of products
by_CustomerID = by_CustomerID.sort_values('Quantity',ascending=False)
by_CustomerID
```

```python
print("highest number of products bought by one customer is",by_CustomerID.iloc[0].loc['Quantity'],"product")
by_CustomerID.iloc[0]
```

```python
#Group df2 by 'Country' to Extract important insights
column_of_interest=['Country','UnitPrice','Quantity']
by_Country = df2[column_of_interest].groupby('Country').sum()
by_Country
```

```python
#Change column 'UnitPrice' to 'Revenue$'
by_Country = by_Country.rename(columns={'UnitPrice':'Revenue$'})
by_Country
```

```python
#sort DataFrame by Revenue to see Country with the highest Revenue
by_Country = by_Country.sort_values('Revenue$',ascending=False)
by_Country
```

```python
# override the default matplotlib style, to avoid the grey background, but keep the grid
sns.set_style("whitegrid") 
#determine the size of the chart
plt.figure(figsize = (12, 6)) 
plt.bar(x = by_Country.index,
       height = by_Country['Quantity'])
plt.xticks( fontsize = 13,rotation=90) # format the labels for the x-axis
plt.yticks(fontsize = 13) # format the y-axis
plt.ylabel('Quantity Sold',fontsize=13)# add label for y-axis
plt.title('Quantity Sold by Countrty',fontsize=18,fontweight='bold')# add title for chart
sns.despine() # removes the top and right border of our graph
plt.show()
```

```python
#Insert new column for Year seperately to use it in Visualization
df2['year'] = df2['InvoiceDate'].apply(lambda x:x.year)
df2['year']
```

```python
#Group df2 by year to Extract important insights and visualize data
column_of_interest = ['year','Quantity', 'UnitPrice']
by_year = df2[column_of_interest].groupby('year').sum()
by_year
```

```python
#Change column 'UnitPrice' to 'Revenue$'
by_year = by_year.rename(columns={'UnitPrice':'Revenue$'})
#add column for Revenue$ in million to visualize correctly
by_year['Revenue$_M'] = by_year['Revenue$']/1000000
by_year
```

```python
# override the default matplotlib style, to avoid the grey background, but keep the grid
sns.set_style("whitegrid") 
#determine the size of the chart
plt.figure(figsize = (12, 6)) 
plt.plot(by_year['Revenue$_M'], linewidth=3)
plt.xticks( ticks=[2010,2011],fontsize = 13) # format the labels for the x-axis
plt.yticks(fontsize = 13) # format the y-axis
plt.ylabel('Revenue in Million Dollars',fontsize=13)# add label for y-axis
plt.title('Revenue Yearly',fontsize=18,fontweight='bold')# add title for chart
sns.despine() # removes the top and right border of our graph
plt.show()
```

```python
#Insert new column for Months seperately to use it in Visualization
df2['month'] = df2['InvoiceDate'].dt.strftime('%Y-%m')
df2['month']
```

```python
#Group df2 by year to Extract important insights and visualize data
column_of_interest = ['month','Quantity', 'UnitPrice']
by_month = df2[column_of_interest].groupby('month').sum()
by_month
```

```python
#Change column 'UnitPrice' to 'Revenue$'
by_month = by_month.rename(columns={'UnitPrice':'Revenue$'})
#add new column for Revenue$ in Thousand to visualize clearly
by_month['Revenue$_K'] = by_month['Revenue$']/1000
by_month
```

```python
#there are missing months that needs to be added in order to visualize data correctly
by_month.loc['2010-04'] =[0,0,0]
by_month.loc['2010-11'] =[0,0,0]
by_month = by_month.sort_index() #sort table by month ascendingly
by_month
```

```python
# override the default matplotlib style, to avoid the grey background, but keep the grid
sns.set_style("whitegrid") 
#determine the size of the chart
plt.figure(figsize = (12, 6)) 
plt.plot(by_month['Revenue$_K'], linewidth=3)
plt.xticks( fontsize = 13 , rotation=90) # format the labels for the x-axis
plt.yticks(fontsize = 13) # format the y-axis
plt.xlabel('YYYY-MM',fontsize=13)
plt.ylabel('Revenue in Thousand Dollar',fontsize=13)# add label for y-axis
plt.title('Revenue Monthly',fontsize=18,fontweight='bold')# add title for chart
sns.despine() # removes the top and right border of our graph
plt.show()
```
